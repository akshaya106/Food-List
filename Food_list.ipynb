```python
import pandas as pd

df = pd.read_csv('/content/FOOD List.csv')
df.head()
```

```python
total_orders = len(df)
print("Total Orders:", total_orders)
```

```python
popular_dishes = df['Food Name'].value_counts()

print(popular_dishes.head(10))
```

```python
city_orders = df['city'].value_counts()

print(city_orders)
```

```python
total_spending = df['Total Spending '].sum()

print("Total Spending:", total_spending)
```

```python
avg_rating = df['Rating'].mean()

print("Average Rating:", round(avg_rating, 2))
```

```python
print("Total Customers:", df['Customer_ID'].nunique())
print("Top Food:", df['Food Name'].mode()[0])
print("Top City:", df['city'].mode()[0])
print("Total Spending:", df['Total Spending '].sum())
print("Average Rating:", round(df['Rating'].mean(), 2))
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```
